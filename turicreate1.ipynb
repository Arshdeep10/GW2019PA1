```
#question 2 filtering the houses acc to sqft living in btween 2000-4000
def sqft_filter(sqft):
  if sqft >2000 and sqft<4001:
    return sqft
  else:
    -1

sf['house_filter'] = sf['sqft_living'].apply(sqft_filter)
a= sf['house_filter'].dropna()
a.explore()
#21.6
```

```
train_data, test_data = sf.random_split(0.8, seed = 0)
```

#Build regression model

```
sqft_model = turicreate.linear_regression.create(train_data, target='price', features=['sqft_living15'])
```

#evaluation of model

```
test_data['price'].mean()
```

```
sqft_model.evaluate(test_data)
```

quite high error and rsme is also too high

```
import matplotlib.pyplot as plt
```

```
plt.plot(test_data['sqft_living15'],test_data['price'],'.',test_data['sqft_living15'],sqft_model.predict(test_data),'-')
```

```
advanced_features = [
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
'condition', # condition of house				
'grade', # measure of quality of construction				
'waterfront', # waterfront property				
'view', # type of view				
'sqft_above', # square feet above ground				
'sqft_basement', # square feet in basement				
'yr_built', # the year built				
'yr_renovated', # the year renovated				
'lat', 'long', # the lat-long of the parcel				
'sqft_living15', # average sq.ft. of 15 nearest neighbors 				
'sqft_lot15', # average lot size of 15 nearest neighbors 
]
advance_model = turicreate.linear_regression.create(train_data,target = 'price', features= advanced_features)
```

```
print(my_features_model.evaluate(test_data))
print(advance_model.evaluate(test_data))
```
